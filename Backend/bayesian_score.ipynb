```python
# load reviews

import pandas as pd

animes = pd.read_parquet(r"C:\Users\insiy\Documents\Projects_Extracurricular\Anime_Recommedation_System\data\animes.parquet")
```

```python
C = animes['score'].mean()        # global mean score
m = animes['members'].quantile(0.75)  # minimum members threshold (75th percentile)

def bayesian_rating(row):
    v = row['members']
    R = row['score']
    return (v / (v + m)) * R + (m / (v + m)) * C

animes['weighted_score'] = animes.apply(bayesian_rating, axis=1)

# Top recommendations
top_animes = animes.sort_values('weighted_score', ascending=False).head(10)
top_animes[['title', 'score', 'members', 'weighted_score']]
```

```python
# Extract base title (text before any special characters like °, ', :)
import re

# Sort first, then drop duplicates keeping only the highest-ranked entry per base title
top_animes = (
    animes.sort_values('weighted_score', ascending=False)
    .drop_duplicates(subset='base_title', keep='first')
)

most_popular = (
    animes.sort_values('popularity')
    .drop_duplicates(subset='base_title',keep='first')
)
```
